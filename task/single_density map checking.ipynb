```python
import h5py
import matplotlib.pyplot as plt

# Specify the exact path to your .h5 file
file_path = r"D:\Data Science\CSRNet-pytorch\ShanghaiTech_Crowd_Counting_Dataset\part_A_final\train_data\ground_truth\IMG_230.h5"  # Replace with your .h5 file path

# Open the .h5 file
with h5py.File(file_path, 'r') as hf:
    # Print the keys in the file to see available datasets
    print("Keys: ", list(hf.keys()))  # This will list available datasets

    # Access the density dataset (change 'density' if your dataset has a different name)
    density_data = hf['density'][:]  # Load the density image

# Visualize the density image
plt.imshow(density_data,cmap="jet")  # You can change the colormap as needed
plt.colorbar()  # Show a colorbar for reference
plt.title('Density Map')  # Title for the plot
plt.axis('off')  # Optionally hide the axis
plt.show()

```
